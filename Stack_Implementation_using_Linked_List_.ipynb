```python
# Stack Implementation using Linked List
# Node Class
class Node:

    def __init__(self, value):
        self.value = value
        self.next_node = None

    def get_value(self):
        return self.value

    def get_next_node(self):
        return self.next_node

    def set_next_node(self, next_node):
        self.next_node = next_node


# Stack Class
class Stack:

    def __init__(self, limit=1000):

        self.top_item = None
        self.size = 0
        self.limit = limit


    # Push Operation
    def push(self, value):

        if self.has_space():

            item = Node(value)

            item.set_next_node(self.top_item)

            self.top_item = item

            self.size += 1

            print(value, "added to stack")

        else:
            print("All out of space!")


    # Pop Operation
    def pop(self):

        if self.size > 0:

            item_to_remove = self.top_item

            self.top_item = item_to_remove.get_next_node()

            self.size -= 1

            return item_to_remove.get_value()

        else:
            print("This stack is totally empty.")


    # Peek Operation
    def peek(self):

        if self.size > 0:
            return self.top_item.get_value()

        else:
            print("Nothing to see here!")


    # Check Empty
    def is_empty(self):
        return self.size == 0


    # Check Space
    def has_space(self):
        return self.size < self.limit


    # Stack Size
    def get_size(self):
        return self.size


# ================= TEST =================

books = Stack()

books.push("Python")
books.push("Java")
books.push("C++")

print("Top Item:", books.peek())

print("Removed:", books.pop())

print("New Top:", books.peek())

print("Stack Size:", books.get_size())
```
